__Задание №3.__ Исследовать функцию:

$$\large f(x) = x^3 - x^2$$

- Область определения $D(f) = (-\infty, +\infty)$.   
  Область значений $E(f) = (-\infty, +\infty)$.


- Нули функции и их кратность:   
  $\normalsize x^3 - x^2 = x^2(x - 1) = 0$   
  $\normalsize x_{1, 2} = 0\,\,\,\,\,x_3 = 1$   
  $\normalsize f(0) = f(1) = 0$


- Отрезки знакопостоянства:   
  $\normalsize f(x) < 0$ при $x \in (-\infty, 0)\cup (0, 1)$,   
  $\normalsize f(x) > 0$ при $x \in (0, \infty)$.


- Интервалы монотонности:   
  $\normalsize f'(x) = 3x^2 - 2x = x(3x - 2) = 0$   
  $\normalsize x_1 = 0,\,\,\,\,\,x_2 = \frac{2}{3}$   
  Функция $f(x)$ возрастает при $f'(x) > 0$ в диапазоне $x \in (-\infty, 0) \cup (\frac{2}{3}, +\infty)$.   
  Функция $f(x)$ убывает при $f'(x) < 0$ в диапазоне $x \in (0, \frac{2}{3})$.


- Функция ни четна, ни нечетна.


- Функция не ограничена.


- Функция непериодическая.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

def func(arg):
    val = pow(arg, 3) - pow(arg, 2)
    return val

x = np.linspace(-0.8, 1.5, 24)
f_x = func(x)

fig = plt.figure()
fig.set_size_inches(4, 6)

plt.xlabel('x')
plt.ylabel('f(x)')

plt.plot(x, f_x)
plt.scatter([0, 1], [0, func(1)], color='green', label='zeroes')
plt.scatter([0, 2/3], [0, func(2/3)], color='red', label='extremums')

plt.legend(loc='upper left')
plt.grid()
plt.show()
```
