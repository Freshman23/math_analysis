### Задание №1.
Исследовать ряд на сходимость, используя признак д’Аламбера:

$\large\displaystyle\sum_{n=1}^{\infty}\frac{n^n}{(n!)^2}$

Для сходимости ряда согласно признака д'Аламбера предел отношения n+1-го и n-го членов ряда должен быть меньше 1 при n, стремящемся к бесконечности.

$\large\displaystyle\lim_{n\to+\infty}\frac{a_{n+1}}{a_n} = \lim_{n\to+\infty}\Bigg[\frac{(n+1)^{n+1}}{((n+1)!)^2}:\frac{n^n}{(n!)^2}\Bigg] = \lim_{n\to+\infty}\Bigg[\frac{(n+1)\cdot(n+1)^n}{(n+1)^2\cdot(n!)^2}\cdot\frac{(n!)^2}{n^n}\Bigg] = \lim_{n\to+\infty}\Bigg[\frac{1}{n+1}\cdot\Bigg(\frac{n+1}{n}\Bigg)^n\Bigg] = \lim_{n\to+\infty}\frac{1}{n+1}\cdot\lim_{n\to+\infty}\Bigg(\frac{n+1}{n}\Bigg)^n = \lim_{n\to+\infty}\frac{1}{n+1}\cdot\lim_{n\to+\infty}\Bigg(1+\frac{1}{n}\Bigg)^n = 0\cdot e = 0 < 1$

Следовательно ряд сходится.

### Задание №2.
Исследовать ряд на сходимость, используя радикальный признак Коши:

$\large\displaystyle\sum_{n=1}^{\infty}\frac{n}{2^n}$

Для сходимости ряда согласно радикального признака Коши предел корня n-ой степени из n-го члена ряда должен быть меньше 1 при n, стремящемся к бесконечности.

$\large\displaystyle\lim_{n\to+\infty}\sqrt[n]{a_n} = \lim_{n\to+\infty}\sqrt[n]{\frac{n}{2^n}} = \lim_{n\to+\infty}\frac{\sqrt[n]{n}}{2} = \frac{1}{2}\lim_{n\to+\infty}\sqrt[n]{n}$

Прологарифмируем последний предел без коэффициента и, воспользовавшись свойствами логарифмов и пределов (в том числе правилом Лопиталя), найдем его значение:

$\large\displaystyle\ln\bigg(\lim_{n\to+\infty}\sqrt[n]{n}\bigg) = \lim_{n\to+\infty}\ln\sqrt[n]{n} = \lim_{n\to+\infty}\ln{n^{\frac{1}{n}}} = \lim_{n\to+\infty}\frac{\ln{n}}{n} = \lim_{n\to+\infty}\frac{(\ln{n})'}{n'} = \lim_{n\to+\infty}\frac{1}{n} = 0$

Тогда

$\large\displaystyle\lim_{n\to+\infty}\sqrt[n]{a_n} = \frac{1}{2}\lim_{n\to+\infty}\sqrt[n]{n} = \frac{1}{2}e^{\ln\big(\lim_{n\to+\infty}\sqrt[n]{n}\big)} = \frac{1}{2}e^0 = \frac{1}{2} < 1$

Следовательно данный ряд сходится.

### Задание №3.
Исследовать ряд на сходимость, используя признак Лейбница:

$\large\displaystyle\sum_{n=1}^{\infty}\frac{(-1)^n}{n+\ln n}$

Проверим ряд на схождение согласно признаку Лейбница:

1. $\large\displaystyle\lim_{n\to+\infty}\frac{(-1)^n}{n+\ln n} = 0$

2. $\large\displaystyle\frac{1}{n+\ln n} > \frac{1}{n+1+\ln(n+1)}$

Значит данный знакочередующийся ряд сходится.

Проверим теперь скорость схождения ряда, для чего исследуем на схождение ряд по модулю:

$\large\displaystyle\sum_{n=1}^{\infty}\frac{1}{n+\ln n}$

Для исследования на схождение этого ряда используем 2-ой признак сравнения и будем сравнивать с гармоническим рядом $\displaystyle\sum_{n=1}^{\infty}\frac{1}{n}$.

Найдем предел отношения n-ых членов сравниваемых рядов:

$\large\displaystyle\lim_{n\to+\infty}\Bigg(\frac{1}{n}:\frac{1}{n+\ln n}\Bigg) = \lim_{n\to+\infty}\frac{n+\ln n}{n} = \lim_{n\to+\infty}\Bigg(1+\frac{\ln n}{n}\Bigg) = 1+\lim_{n\to+\infty}\frac{\ln n}{n} = 1+\lim_{n\to+\infty}\frac{(\ln n)'}{n'} = 1+\lim_{n\to+\infty}\frac{1}{n} = 1$

Предел отношения конечное число, следовательно так как расходится гармонический ряд, то расходится и ряд $\displaystyle\sum_{n=1}^{\infty}\frac{1}{n+\ln n}$.

А значит данный знакочередующийся ряд $\displaystyle\sum_{n=1}^{\infty}\frac{(-1)^n}{n+\ln n}$ сходится условно.

### Задание №4.
Исследовать ряд на сходимость, используя признак Раабе:

$\large\displaystyle\sum_{n=1}^{\infty}\frac{3^n}{2^n}$

Этот ряд в принципе не сходится из-за невыполнения необходимого условия сходимости:

$\large\displaystyle\lim_{n\to+\infty}\frac{3^n}{2^n} = \lim_{n\to+\infty}\Bigg(\frac{3}{2}\Bigg)^n = +\infty$

Но проверим на сходимость по признаку Раабе:

$\large\displaystyle\lim_{n\to+\infty}n\Bigg(\frac{3^n}{2^n}:\frac{3^{n+1}}{2^{n+1}}-1\Bigg) = \lim_{n\to+\infty}n\Bigg(\frac{3^n}{2^n}\cdot\frac{2^n\cdot2}{3^n\cdot3}-1\Bigg) = \lim_{n\to+\infty}n\Bigg(\frac{2}{3}-1\Bigg) = -\infty < 1$

Следовательно ряд расходится.

### Задание №5.
Разложить функцию по Тейлору в единице:

$\displaystyle\large f(x) = \ln(16x^2)$

Воспользуемся свойства логарифмов, чтобы преобразовать функцию для упрощения дальнейшего дифференцирования.

$\displaystyle\large f(x) = \ln(16x^2) = 4\ln2 +2\ln x$

Общая формула для разложения в ряд Тейлора:

$\displaystyle\large f(x) = \sum_{n=0}^{+\infty}\frac{f^{(n)}(a)}{n!}(x-a)^n$, где a=1 (в данном примере).

Найдем значения функции и первых её производных в заданной точке (а=1):

$\displaystyle\large f(1) = 4\ln2$

$\displaystyle\large f'(x) = \frac{2}{x}\Bigg|_{x=1} = 2$

$\displaystyle\large f''(x) = -\frac{2}{x^2}\Bigg|_{x=1} = -2$

$\displaystyle\large f'''(x) = \frac{4}{x^3}\Bigg|_{x=1} = 4$

$\displaystyle\large f^{(4)}(x) = -\frac{12}{x^4}\Bigg|_{x=1} = -12$

$\displaystyle\large f^{(5)}(x) = \frac{48}{x^5}\Bigg|_{x=1} = 48$

$\displaystyle\large f^{(6)}(x) = -\frac{240}{x^6}\Bigg|_{x=1} = -240$

$\displaystyle\large f^{(n)}(1) = 2\cdot(-1)^{n-1}\cdot(n-1)!$

Подставим в формулу:

$\displaystyle\large \ln(16x^2) = 4\ln2 + \sum_{n=1}^{+\infty}\frac{2\cdot(-1)^{n-1}\cdot(n-1)!}{n!}(x-1)^n = 4\ln2 + 2\sum_{n=1}^{+\infty}\frac{(-1)^{n-1}\cdot(n-1)\cdot n!}{n!}(x-1)^n = 4\ln2 + 2\sum_{n=1}^{+\infty}(-1)^{n}\cdot(1-n)\cdot(x-1)^n = 4\ln2 + 2\sum_{n=1}^{+\infty}\bigg[(1-n)(1-x)^n\bigg]$

### Задание №6.
Дана функция $\large f(x)=x^2$
  - Разложить функцию в ряд Фурье по косинусам на отрезке x ∈ $[-\pi;\pi]$;
  - Построить график функции и ее разложения.
  
Общая формула разложения функции в ряд Фурье на отрезке $[-l,l]$:

$\large\displaystyle f(x) \sim \frac{a_0}{2}+\sum_{n=1}^{\infty}a_n\cos\frac{\pi nx}{l}+b_n\sin\frac{\pi nx}{l}$, где

$\large\displaystyle a_0 = \frac{1}{l}\int_{-l}^{l}f(x)dx$,

$\large\displaystyle a_n = \frac{1}{l}\int_{-l}^{l}f(x)\cos\frac{\pi nx}{l}dx\,\,\,\,\,(n\in\mathbb N)$,

$\large\displaystyle b_n = \frac{1}{l}\int_{-l}^{l}f(x)\sin\frac{\pi nx}{l}dx\,\,\,\,\,(n\in\mathbb N)$.

Для нашего случая четности функции $x^2$ и отрезка разложения $[-\pi;\pi]$ формулы нахождения коэффициентов Фурье примут вид:

$\large\displaystyle a_0 = \frac{2}{\pi}\int_{0}^{\pi}f(x)dx$,

$\large\displaystyle a_n = \frac{2}{\pi}\int_{0}^{\pi}f(x)\cos(nx)dx\,\,\,\,\,(n\in\mathbb N)$,

$\large\displaystyle b_n = 0$.

А формула разложения функции:

$\large\displaystyle f(x) \sim \frac{a_0}{2}+\sum_{n=1}^{\infty}a_n\cos(nx)$

Разложим функцию:

$\large\displaystyle a_0 = \frac{2}{\pi}\int_{0}^{\pi}x^2dx = \frac{2x^3}{3\pi}\Bigg|_{0}^{\pi} = \frac{2}{3}\pi^2$

$\large\displaystyle a_n = \frac{2}{\pi}\int_{0}^{\pi}x^2\cos(nx)dx = (*)$

Проинтегрируем по частям:

$\large\displaystyle U=x^2\,\,\,\Rightarrow\,\,\,dU=2xdx$

$\large\displaystyle dV=\cos(nx)dx\,\,\,\Rightarrow\,\,\,V=\frac{1}{n}\sin(nx)$

Тогда:

$\large\displaystyle (*) = \frac{2}{\pi}\Bigg[\frac{x^2}{n}\sin(nx)\Bigg|_0^\pi - \frac{2}{n}\int_{0}^{\pi}x\sin(nx)dx\Bigg] = \frac{2}{\pi}\Bigg(\frac{\pi^2}{n}\sin(n\pi)-\frac{0}{n}\sin0\Bigg)-\frac{4}{n\pi}\int_{0}^{\pi}x\sin(nx)dx = -\frac{4}{n\pi}\int_{0}^{\pi}x\sin(nx)dx = (*)$

Проинтегрируем по частям еще раз:

$\large\displaystyle U=x\,\,\,\Rightarrow\,\,\,dU=dx$

$\large\displaystyle dV=-\sin(nx)dx\,\,\,\Rightarrow\,\,\,V=\frac{1}{n}\cos(nx)$

Получаем:

$\large\displaystyle (*) = -\frac{4}{n\pi}\int_{0}^{\pi}x\sin(nx)dx = \frac{4}{n\pi}\Bigg[\frac{x}{n}\cos(nx)\Bigg|_0^\pi-\frac{1}{n}\int_0^\pi\cos(nx)dx\Bigg] = \frac{4\pi}{n^2\pi}\cos(n\pi) - \frac{4}{n^3\pi}\sin(nx)\Bigg|_0^\pi = \frac{4}{n^2}\cos(n\pi) = \frac{4\cdot(-1)^n}{n^2}$

Таким образом:

$\large\displaystyle x^2 \sim \frac{\pi^2}{3} + 4\sum_{n=1}^{\infty}\frac{(-1)^n}{n^2}\cos(nx)$

```python
from sympy import *
from sympy.plotting import plot
init_printing()
x = Symbol('x')

def fourier_series_xpow2(num):
    x = Symbol('x')
    series = pi**2/3
    for n in range(1, num+1):
        series += 4*(-1)**n*cos(n*x)/n**2
    return series

p1 = plot(x**2, (x, -4, 4), label='x^2', line_color='blue', show=False)
p2 = plot(fourier_series_xpow2(3), (x, -4*pi, 4*pi), label='fourier_series_3', legend=True, line_color='red', size=(8, 4), show=False)
p3 = plot(fourier_series_xpow2(7), (x, -4*pi, 4*pi), label='fourier_series_7', legend=True, line_color='red', size=(8, 4), show=False)
p4 = plot(fourier_series_xpow2(10), (x, -4*pi, 4*pi), label='fourier_series_10', legend=True, line_color='red', size=(8, 4), show=False)
p2.extend(p1)
p3.extend(p1)
p4.extend(p1)
p2.show()
p3.show()
p4.show()
```

Ряд Фурье представляет собой периодическую функцию, которая в интервале разложения (в данном случае $[-\pi,\pi]$) приближается к раскладываемой функции по мере увеличения количества членов ряда.
