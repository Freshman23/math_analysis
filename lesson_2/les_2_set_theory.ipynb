1. Прочитать высказывания математической логики, построить их отрицания и установить истинность.

- $\forall y \in [0; 1] : sgn(y) = 1$   
Для любого $y$ в диапазоне от 0 включительно до 1 включительно функция $sgn(y)$ равна 1.   
__Решение__   
$$sgn(y) = \begin{cases}
1 & if & y>0, \\
0 & if & y=0, \\
-1 & if & y<0. \end{cases}$$
Так как по определению при $y=0$: $sgn(y)=0$, то данное высказывание ложно.   
Отрицание данного высказывания: $\exists y \in [0; 1] : sgn(y) \neq 1$.   
Существует такой $y$ из диапазона от 0 включительно до 1 включительно, для которого функция $sgn(y)$ не равна 1. Что, исходя из определения, верно. Следовательно данное высказывание ложно.   
__Ответ__   
Данное высказывание ложно.
___
- $\forall n \in \mathbb N > 2 : \exists x, y, z \in \mathbb N : x^n + y^n = z^n$   
Для любого $n$, принадлежащего множеству натуральных чисел, больших чем 2, существуют $x, y, z$, принадлежащие множеству натуральных чисел, такие, что выполняется равенство $x^n + y^n = z^n$.   
__Решение__   
Согласно доказанной теореме Ферма при любом натуральном числе $n > 2$ уравнение: $x^n + y^n = z^n$ не имеет решений в натуральных числах $x, y, z$. Данное высказывание ложно.   
Отрицание данного высказывания: $\exists n \in \mathbb N > 2 : \forall x,y,z \in \mathbb N : x^n + y^n \neq z^n$.   
Для любых натуральных $x,y,z$ существует такое натуральное число $n$, большее 2, при котором $x^n + y^n \neq z^n$. Что истинно, также в следствие доказательства теоремы Ферма.   
__Ответ__   
Данное высказывание ложно.
___
- $\forall x \in \mathbb R: \exists X \in \mathbb R: X>x$   
Для любого вещественного числа $x$ существует такое вещественное число $X$, при котором $X > x$.   
__Решение__   
Множества вещественных чисел - бесконечные. По определению бесконечного множества: для каждого вещественного числа найдется большее его вещественное число. Следовательно данное высказывание истинно.   
Отрицание данного высказывания: $\exists x \in \mathbb R: \nexists X \in \mathbb R: X \leq x$   
Существует такое вещественное число $x$, для которого не найдется другого вещественного числа $X$, чтобы $X \leq x$.   
Так как множество вещественных чисел бесконечно и в отрицательную сторону, следовательно нет такого вещественного числа, которому нельзя было бы найти меньшее либо равное ему вещественное число. Отрицание данного высказывания ложно.   
__Ответ__   
Данное высказывание истинно.
___
- $\forall x \in \mathbb C: \nexists y \in \mathbb C: x>y \parallel x<y$   
Для любого комплексного числа $x$ не существует такого комплексного числа $y$, при котором $x>y$ либо $x<y$.   
__Решение__   
Очевидно, что на бесконечной плоскости комплексных чисел для любого комплесного числа найдется такое комплексное число, которое будет больше или меньше первого. Данное высказывание ложно.   
Отрицание данного высказывания: $\exists x, y \in \mathbb C: x = y$   
Существуют такие комплексные числа $x, y$, которые будут равны между собой. Очевидно, что отрицание высказывания истинно.   
__Ответ__   
Данное высказывание ложно.
___
- $\forall y \in [0; \frac{\pi}{2}]: \exists \varepsilon > 0: \sin y < \sin (y + \varepsilon)$   
Для всех $y$ из множетсва чисел на отрезке от 0 включительно до $\frac{\pi}{2}$ включительно существует такое $\varepsilon$, большее 0, при котором выполняется неравенство: $\sin y < \sin (y + \varepsilon)$.   
__Решение__   
Область значений функции $\sin y \in [-1; 1]$. При $y \in [0; \frac{\pi}{2}]$ значения функции $\sin y \in [0; 1]$. В частности $\sin \frac{\pi}{2} = 1$ и из области значений функции не найдется такого числа, которое было бы больше 1. Следовательно и не существует такого $\varepsilon$, при котором было бы верно равенство: $\sin \frac{\pi}{2} < sin (\frac{\pi}{2} + \varepsilon)$. Данное высказывание ложно.   
Отрицание данного высказывания: $\exists y \in [0; \frac{\pi}{2}]: \forall \varepsilon > 0: \sin y \geq \sin (y + \varepsilon)$   
Для любого $\varepsilon$, большего 0, существует такой $y$ из множетсва чисел на отрезке от 0 включительно до $\frac{\pi}{2}$ включительно, что выполняется неравенство $\sin y \geq \sin (y + \varepsilon)$.   
Опять же исходя из области значений функции $\sin y \in [-1; 1]$ при любом $\varepsilon$, большим 0, всегда будет выполнятся неравенство: $\sin \frac{\pi}{2} \geq \sin (\frac{\pi}{2} + \varepsilon)$. То есть $\sin (\frac{\pi}{2} + \varepsilon) \leq 1$. Отрицание данного высказывания истинно.   
__Ответ__   
Данное высказывание ложно.
___
- $\forall y \in [0; \pi): \exists \varepsilon > 0: \cos y > \cos (y + \varepsilon)$   
Для всех $y$ из множетсва чисел на отрезке от 0 включительно до $\pi$ исключительно существует такое $\varepsilon$, большее 0, при котором выполняется неравенство: $\cos y > \cos (y + \varepsilon)$.   
__Решение__   
Область значений функции $\cos y \in [-1; 1]$. При $y \in [0; \pi)$ значения функции $\cos y \in (-1; 1]$. Для любого $y \in [0; \pi)$ мы можем подобрать значение $\varepsilon$ такое, чтобы $y + \varepsilon = \pi$ и $\cos \pi = -1$. Следовательно и выполняется неравенство: $\cos y > -1$ при $y \in [0; \pi)$. Данное высказывание истинно.   
Отрицание данного высказывания: $\exists y \in [0; \pi): \forall \varepsilon > 0: \cos y \leq \cos (y + \varepsilon)$   
Для любого $\varepsilon$, большего 0, существует такой $y$ из множетсва чисел на отрезке от 0 включительно до $\pi$ исключительно, что выполняется неравенство: $\cos y \leq \cos (y + \varepsilon)$.   
Для любого $y \in [0; \pi)$ мы можем подобрать значение $\varepsilon$ такое, чтобы $y + \varepsilon = \pi$ и $\cos \pi = -1$. Следовательно и не будет выполняться неравенство: $\cos y \leq -1$ при $y \in [0; \pi)$. Отрицание данного высказывания ложно.   
__Ответ__   
Данное высказывание истинно.

Также если рассматривать $x$ как число, то любой $x$ принадлежит множеству, состоящему из всех числовых множеств. Отрицание данного высказывание истинно.   
__Ответ__   
Данное высказывание ложно.

Пусть:   
$A = \{1,\,2,\,3,\,one,\,two,\,three\}$   
$B = \{3,\,4,\,5,\,three,\,four,\,five\}$   
$C = \{1,\,2,\,3,\,three,\,four,\,five\}$   
Тогда:   
$A \cup B \cup C = \{1,\,2,\,3,\,4,\,5,\,one,\,two,\,three,\,four,\,five\}$   
$A \cap B \cap C = \{3,\,three\}$   
$(A - B) - C = \{one,\,two\}$    
$A \Delta B \Delta C = \{3,\,4,\,5,\,one,\,two,\,three\}$   
От перемены мест множеств и порядка выполнения операции изменяется только разность:   
$A - (B - C) = \{1,\,2,\,3,\,one,\,two,\,three\}$   
$B - (C - A) = \{3,\,4,\,5,\,three\}$
