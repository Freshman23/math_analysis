# Тема "Функции нескольких переменных"

### Задание №1.   
Исследовать функцию на условный экстремум.

$\normalsize U=3-8x+6y$, если $\normalsize x^2+y^2=36$

Составим функцию Лагранжа:

$\large L(x,y,\lambda) = 3-8x+6y+\lambda\cdot(x^2+y^2-36)$

___Необходимое условие экстремумов.___   
Найдем частные производные 1-го порядка, приравняем их к нулю и решим систему уравнений:

$\large
 \begin{cases}
    L'_x = -8+\lambda\cdot2x = 0 \\
    L'_y = 6+\lambda\cdot2y = 0 \\
    L'_\lambda = x^2+y^2-36 = 0
  \end{cases} \Rightarrow
 \begin{cases}
    x = \displaystyle\frac{4}{\lambda} \\
    y = \displaystyle-\frac{3}{\lambda} \\
    \Bigg(\displaystyle\frac{4}{\lambda}\Bigg)^2+\Bigg(\displaystyle-\frac{3}{\lambda}\Bigg)^2 = 36
  \end{cases} \Rightarrow
 \begin{cases}
    x = \displaystyle\frac{4}{\lambda} \\
    y = \displaystyle-\frac{3}{\lambda} \\
    \lambda^2 = \displaystyle\frac{25}{36}
  \end{cases} \Rightarrow
 \begin{cases}
    x = \displaystyle\frac{4}{\lambda} \\
    y = \displaystyle-\frac{3}{\lambda} \\
    \lambda = \displaystyle\pm\frac{5}{6}
  \end{cases}$
  
Получаем две точки: $\displaystyle\normalsize \bigg(-\frac{5}{6}, -4.8, 3.6\bigg), \bigg(\frac{5}{6}, 4.8, -3.6\bigg)$

___Достаточное условие экстремумов.___   
Найдем частные производные 2-го порядка:

$\large L''_{\lambda\lambda} = 0$

$\large L''_{xx} = 2\lambda$

$\large L''_{yy} = 2\lambda$

$\large L''_{xy} = L''_{yx} = 0$

$\large L''_{\lambda x} = L''_{x\lambda} = 2x$

$\large L''_{\lambda y} = L''_{y\lambda} = 2y$

Составим матрицу Гёссе:

$\large\begin{pmatrix}
    L''_{\lambda\lambda} & L''_{\lambda x} & \large L''_{\lambda y} \\
    L''_{x\lambda} & L''_{xx} & \large L''_{xy} \\
    L''_{y\lambda} & L''_{yx} & \large L''_{yy}
  \end{pmatrix} \Rightarrow 
 \large\begin{pmatrix}
    0 & 2x & 2y \\
    2x & 2\lambda & 0 \\
    2y & 0 & 2\lambda
  \end{pmatrix}$

Найдем определитель матрицы Гёссе в общем виде:

$\large\Delta = \begin{vmatrix}
    0 & 2x & 2y \\
    2x & 2\lambda & 0 \\
    2y & 0 & 2\lambda
  \end{vmatrix} =
 (0\cdot2\lambda\cdot2\lambda+0\cdot2x\cdot2y+0\cdot2x\cdot2y)-(2y\cdot2y\cdot2\lambda+2x\cdot2x\cdot2\lambda+0\cdot0\cdot0) = -8\lambda(x^2+y^2)$

Отсюда получаем, что $\normalsize\Delta<0$ при $\normalsize\lambda>0$ и это точка минимума; $\normalsize\Delta>0$ при $\normalsize\lambda<0$ и это точка максимума.

__Ответ.__ $\displaystyle\normalsize \bigg(-\frac{5}{6}, -4.8, 3.6\bigg)$ - точка максимума,
$\displaystyle\normalsize \bigg(\frac{5}{6}, 4.8, -3.6\bigg)$ - точка минимума.
___

### Задание №2.   
Исследовать функцию на условный экстремум.

$\normalsize U=2x^2+12xy+32y^2+15$, если $\normalsize x^2+16y^2=64$

Составим функцию Лагранжа:

$\large L(x,y,\lambda) = 2x^2+12xy+32y^2+15+\lambda\cdot(x^2+16y^2-64)$

___Необходимое условие экстремумов.___   
Найдем частные производные 1-го порядка, приравняем их к нулю и решим систему уравнений:

$\large
 \begin{cases}
    L'_x = 4x+12y+\lambda\cdot2x = 0 \\
    L'_y = 12x+64y+\lambda\cdot32y = 0 \\
    L'_\lambda = x^2+16y^2-64 = 0
  \end{cases} \Rightarrow
 \begin{cases}
    12y+2x\cdot(2+\lambda) = 0 \\
    12x+32y\cdot(2+\lambda) = 0 \\
    x^2+16y^2-64 = 0
  \end{cases} \Rightarrow
 \begin{cases}
    2+\lambda = \displaystyle-\frac{6y}{x} \\
    2+\lambda = \displaystyle-\frac{3x}{8y} \\
    x^2+16y^2-64 = 0
  \end{cases}$

Приравняв правые половины первых двух уравнений, получим:

$\displaystyle\large\frac{6y}{x}=\frac{3x}{8y}\,\Rightarrow\,3x^2=48y^2\,\Rightarrow\,x^2=16y^2$

Подставим полученную зависимость в третье уравение системы уравнений:

$\displaystyle\large x^2+x^2-64=0\,\Rightarrow\,x^2=32\,\Rightarrow\,x=\pm4\sqrt{2}$

Тогда

$\displaystyle\large 32=16y^2\,\Rightarrow\,y=\pm\sqrt{2}$

Подставив каждую комбинацию аргументов $x$ и $y$ в одно из первых двух уравнений системы уравнений и найдя соответствующее значение $\lambda$, получим четыре точки:

$\displaystyle\normalsize \bigg(-3.5, -4\sqrt{2}, -\sqrt{2}\bigg), \bigg(-3.5, 4\sqrt{2}, \sqrt{2}\bigg), \bigg(-0.5, -4\sqrt{2}, \sqrt{2}\bigg), \bigg(-0.5, 4\sqrt{2}, -\sqrt{2}\bigg)$

___Достаточное условие экстремумов.___   
Найдем частные производные 2-го порядка:

$\large L''_{\lambda\lambda} = 0$

$\large L''_{xx} = 2\lambda+4$

$\large L''_{yy} = 32\lambda+64$

$\large L''_{xy} = L''_{yx} = 12$

$\large L''_{\lambda x} = L''_{x\lambda} = 2x$

$\large L''_{\lambda y} = L''_{y\lambda} = 32y$

Составим матрицу Гёссе:

$\large\begin{pmatrix}
    L''_{\lambda\lambda} & L''_{\lambda x} & \large L''_{\lambda y} \\
    L''_{x\lambda} & L''_{xx} & \large L''_{xy} \\
    L''_{y\lambda} & L''_{yx} & \large L''_{yy}
  \end{pmatrix} \Rightarrow 
 \large\begin{pmatrix}
    0 & 2x & 32y \\
    2x & 2\lambda+4 & 12 \\
    32y & 12 & 32\lambda+64
  \end{pmatrix}$

Найдем определитель матрицы Гёссе в общем виде:

$\large\Delta = \begin{vmatrix}
    0 & 2x & 32y \\
    2x & 2\lambda+4 & 12 \\
    32y & 12 & 32\lambda+64
  \end{vmatrix} =
 (0\cdot(2\lambda+4)\cdot(32\lambda+64)+12\cdot2x\cdot32y+12\cdot2x\cdot32y)-(32y\cdot32y\cdot(2\lambda+4)+0\cdot12\cdot12+2x\cdot2x\cdot(32\lambda+64)) = 1536xy-(128x^2\cdot(\lambda+2)+2048y^2\cdot(\lambda+2)) = 1536xy-128(\lambda+2)(x^2+16y^2)$

Последний множитель нам известен из условия: $\normalsize x^2+16y^2=64$, поэтому

$\large\Delta = 1536xy-128(\lambda+2)\cdot64 = 1536xy-8192(\lambda+2)$

Также для исследования $\Delta$ относительно 0 можем избавится от больших коэффициентов, разделив выражение на наибольшее общее кратное этиx коэффициентов:

$\large \tilde{\Delta} = 3xy-16(\lambda+2)$

Подставим значения каждой точки:

1. $\large\big(-3.5, -4\sqrt{2}, -\sqrt{2}\big)$

   $\large\tilde{\Delta}_1 = 48 > 0$


2. $\large\big(-3.5, 4\sqrt{2}, \sqrt{2}\big)$

   $\large\tilde{\Delta}_2 = 48 > 0$


3. $\large\big(-0.5, -4\sqrt{2}, \sqrt{2}\big)$

   $\large\tilde{\Delta}_3 = -48 < 0$


4. $\large\big(-0.5, 4\sqrt{2}, -\sqrt{2}\big)$

   $\large\tilde{\Delta}_4 = -48 < 0$

__Ответ:__   
$\large\big(-3.5,\,-4\sqrt{2},\,-\sqrt{2}\big),\,\big(-3.5,\,4\sqrt{2},\,\sqrt{2}\big)$ - точки максимума   
$\large\big(-0.5,\,-4\sqrt{2},\,\sqrt{2}\big),\,\big(-0.5,\,4\sqrt{2},\,-\sqrt{2}\big)$ - точки минимума

### Задание №3.   
Найти производную функции $\normalsize U=x^2+y^2+z^2$ по направлению вектора $\normalsize \vec{c}=(-9,8,-12)$ в точке $\normalsize M_0(8, -12, 9)$.

Найдем частные производные 1-го порядка в точке $M_0$:

$\large U'_x = 2x\,\bigg|_{M_0} = 16$

$\large U'_y = 2y\,\bigg|_{M_0} = -24$

$\large U'_z = 2z\,\bigg|_{M_0} = 18$

Нормализуем вектор $\vec{c}$:

$\large\displaystyle |\vec{c}| = \sqrt{c_x^2+c_y^2+c_z^2} = \sqrt{(-9)^2+8^2+(-12)^2} = 17$

$\large\displaystyle \vec{c_n} = \Bigg(\frac{c_x}{|\vec{c}|}, \frac{c_y}{|\vec{c}|}, \frac{c_z}{|\vec{c}|}\Bigg) = \Bigg(-\frac{9}{17}, \frac{8}{17}, -\frac{12}{17}\Bigg)$

Тогда

$\large\displaystyle U'_{\vec{c}}\bigg|_{M_0} = U'_x\cdot c_{nx} + U'_y\cdot c_{ny} + U'_z\cdot c_{nz} = 16\cdot\Bigg(-\frac{9}{17}\Bigg) - 24\cdot\frac{8}{17} + 18\cdot\Bigg(-\frac{12}{17}\Bigg) = -\frac{552}{17} \approx -32,5$

__Ответ:__ $\large\displaystyle -\frac{552}{17}$
___

### Задание №4.
Найти производную функции $\normalsize U=e^{x^2+y^2+z^2}$ по направлению вектора $\vec{d}=(4,-13,-16)$ в точке $M_0(-16, 4, -13)$.

Найдем частные производные 1-го порядка в точке $M_0$:

$\large U'_x = 2x\cdot e^{x^2+y^2+z^2}\,\bigg|_{M_0} = -32e^{441}$

$\large U'_y = 2y\cdot e^{x^2+y^2+z^2}\,\bigg|_{M_0} = 8e^{441}$

$\large U'_z = 2z\cdot e^{x^2+y^2+z^2}\,\bigg|_{M_0} = -26e^{441}$

Нормализуем вектор $\vec{d}$:

$\large\displaystyle |\vec{d}| = \sqrt{d_x^2+d_y^2+d_z^2} = \sqrt{4^2+(-13)^2+(-16)^2} = 21$

$\large\displaystyle \vec{d_n} = \Bigg(\frac{d_x}{|\vec{d}|}, \frac{d_y}{|\vec{d}|}, \frac{d_z}{|\vec{d}|}\Bigg) = \Bigg(\frac{4}{21}, -\frac{13}{21}, -\frac{16}{21}\Bigg)$

Тогда

$\large\displaystyle U'_{\vec{d}}\bigg|_{M_0} = U'_x\cdot d_{nx} + U'_y\cdot d_{ny} + U'_z\cdot d_{nz} = -32e^{441}\cdot\frac{4}{21} + 8e^{441}\cdot\Bigg(-\frac{13}{21}\Bigg) - 26e^{441}\cdot\Bigg(-\frac{16}{21}\Bigg) = \frac{184}{21}e^{441}$

__Ответ:__ $\large\displaystyle \frac{184}{21}e^{441}$
